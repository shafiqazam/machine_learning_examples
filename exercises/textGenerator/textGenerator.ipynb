```python
import numpy as np
import string
```

```python
initial = {}
firstOrder = {}
secondOrder = {}
word2idx = {}
```

```python
input_texts = []

for line in open('robert_frost.txt', encoding='utf-8'):
        # Remove new line by rstrip
        line = line.rstrip().lower()
        if line:
            # Remove punctuation
            line = line.translate(str.maketrans('','', string.punctuation))
            input_texts.append(line)
```

```python
def addIteminDict(d, k ,v):
    if k not in d:
        d[k] = []
    d[k].append(v)
```

```python
for line in input_texts:
    tokens = line.rstrip().lower().split()
    for idx, token in enumerate(tokens):
        
        if token not in word2idx:
            word2idx[token] = 1
            
        # print(idx)
        # print(token)
        if idx == 0: 
            initial[token] = initial.get(token, 0.) + 1
        else:
            t_1 = tokens[idx - 1]
            # print(t_1)
            # Do for the end line
            if idx == len(tokens) - 1:
                addIteminDict(secondOrder, (t_1, token), 'END')
            # Then preprocess the current idx
            if idx == 1: 
                addIteminDict(firstOrder, t_1, token)
            else:
                t_2 = tokens[idx - 2]
                addIteminDict(secondOrder, (t_2, t_1), token)
            
```

```python
total = sum(initial.values())
for w, c in initial.items():
    initial[w] = c / total
```

```python
def list2probDict(d):
    
    for k, i in d.items():
        dCount = {}
        l = len(i)
        # print(k)
        # print(i)
        for w in i:
            if w not in dCount:
                dCount[w] = 1
            else: 
                dCount[w] += 1
        for w, c in dCount.items():
            dCount[w] = c / l
        d[k] = dCount
    # print("Test")
        
```

```python
list2probDict(firstOrder)
list2probDict(secondOrder)
```

```python
def sample_word(d):
  # print "d:", d
  p0 = np.random.random()
  # print "p0:", p0
  cumulative = 0
  for t, p in d.items():
    cumulative += p
    if p0 < cumulative:
      return t
  assert(False) # should never get here
```

```python
def generate():
  for i in range(4): # generate 4 lines
    sentence = []

    # initial word
    w0 = sample_word(initial)
    sentence.append(w0)

    # sample second word
    w1 = sample_word(firstOrder[w0])
    sentence.append(w1)

    # second-order transitions until END
    while True:
      w2 = sample_word(secondOrder[(w0, w1)])
      if w2 == 'END':
        break
      sentence.append(w2)
      w0 = w1
      w1 = w2
    print(' '.join(sentence))
```

```python
generate()
```
